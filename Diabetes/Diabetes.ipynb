```python
diabetic_data = df.copy()
```

```python
diabetic_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = diabetic_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NAN)
```

```python
p=diabetic_data.hist(figsize=(20,20),bins =20)
plt.plot()
```

```python
diabetic_data['SkinThickness'].fillna(diabetic_data['SkinThickness'].mean(),inplace = True)
```

```python
diabetic_data['BMI'].fillna(diabetic_data['BMI'].mean(),inplace = True)
```

```python
diabetic_data['Glucose'].fillna(diabetic_data['Glucose'].mean(),inplace = True)
diabetic_data['Insulin'].fillna(diabetic_data['Insulin'].mean(),inplace = True)
diabetic_data['BloodPressure'].fillna(diabetic_data['BloodPressure'].mean(),inplace = True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X =  pd.DataFrame(scaler.fit_transform(diabetic_data.drop(["Outcome"],axis = 1),),columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
X.head()
```

```python
target = diabetic_data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,target,test_size = 0.3,random_state = 42 )
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
from sklearn.externals import joblib 
joblib.dump(rf, 'filename.pkl') 
```

```python
model = joblib.load('filename.pkl')
```

```python
rf_train = model.predict(X_train)
```

```python
values = rf.feature_importances_
importance = pd.DataFrame(values).T
importance.columns = X.columns
importance.plot(figsize = (10,10),kind='bar')
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(rf_train,y_train)
```

```python
rf_preds = model.predict(X_test)
```

```python
accuracy_score(y_test,rf_preds)
```

```python
confusion_matrix(y_test,rf_preds)
```
